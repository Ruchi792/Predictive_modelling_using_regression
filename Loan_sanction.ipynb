```python
# prompt: Define the dependent variable y as loan status and rest all independent variable x

# Define dependent and independent variables
y = data['Loan_Status']  # Dependent variable
X = data.drop('Loan_Status', axis=1)  # Independent variables
```

```python
# prompt: extract logistic regreession equation using stats model

import statsmodels.api as sm

# Add a constant to the independent variables
X = sm.add_constant(X)

# Fit the logistic regression model
model = sm.Logit(y, X).fit()

# Print the model summary
print(model.summary())

# Extract the coefficients and intercept
coefficients = model.params
intercept = coefficients['const']
other_coefficients = coefficients.drop('const')

# Construct the logistic regression equation string
equation = "log(odds) = " + str(round(intercept, 3))

for feature, coef in other_coefficients.items():
    equation += " + " + str(round(coef, 3)) + "*" + feature

print("\nLogistic Regression Equation:")
equation
```

Credit History (p = 0.000, coef = 3.8829) is the strongest predictor of loan approval.

Married status (coef = 0.4609) slightly increases approval chances.

Education (-0.4460) & Loan Amount (-0.0002) have minor negative impacts.

Income, Gender, Dependents, and Property Area are not significant predictors.

Model is statistically significant (p < 0.05) but has moderate explanatory power (Pseudo R² = 0.2658).

```python
# prompt:  generate confusion matrix on data

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'model' is your fitted logistic regression model and 'X' and 'y' are your data
# Make predictions on the data
y_pred = model.predict(X)

# Convert probabilities to class labels (0 or 1)
y_pred_class = (y_pred > 0.5).astype(int)


# Create the confusion matrix
cm = confusion_matrix(y, y_pred_class)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

True Positives (TP): 378 (Correctly predicted as 1)

True Negatives (TN): 82 (Correctly predicted as 0)

False Positives (FP): 97 (Incorrectly predicted as 1)

False Negatives (FN): 7 (Incorrectly predicted as 0)



```python
# prompt: drop insignificant independent variable   based on the highest p-value until all variables are significant

import pandas as pd
import numpy as np
import statsmodels.api as sm

def drop_insignificant_variables(y, X):
    # Add a constant to the independent variables
    X = sm.add_constant(X)

    # Keep track of dropped columns
    dropped_columns = []

    while True:
        # Fit the logistic regression model
        model = sm.Logit(y, X).fit(disp=0)  # Suppress iteration output

        # Get p-values for each coefficient
        p_values = model.pvalues

        # Identify the variable with the highest p-value
        max_p_value = p_values.drop('const').max()

        if max_p_value > 0.05 :
            # Find the name of the variable with the maximum p-value
            variable_to_drop = p_values.drop('const').idxmax()

            # Drop the variable with the highest p-value
            X = X.drop(variable_to_drop, axis=1)
            dropped_columns.append(variable_to_drop)

        else:
            break

    print("Dropped columns:", dropped_columns)
    return model, X

# Assuming 'data' DataFrame is already loaded and preprocessed
# ... (your existing code for data loading and preprocessing)

# Define dependent and independent variables
y = data['Loan_Status']
X = data.drop('Loan_Status', axis=1)

# Call the function to drop insignificant variables
final_model, final_X = drop_insignificant_variables(y,X)

# Print the final model summary
print(final_model.summary())
```

Key Predictors:

Credit History (coef = 3.869, p = 0.000): Strongest predictor of loan approval.

Married (coef = 0.509, p = 0.023): Slight positive impact on loan approval.

Intercept (-2.811, p = 0.000): Baseline log-odds of loan approval when all predictors are zero.

Dropped Variables:

Self Employed, Income, Loan Amount, Education, etc. were removed due to insignificance in previous models.

```python
# prompt: Generate confusu=ion matrix

import matplotlib.pyplot as plt
# Assuming 'final_model' and 'final_X' are defined from the previous code.

# Make predictions on the data using the final model
y_pred_final = final_model.predict(final_X)

# Convert probabilities to class labels (0 or 1)
y_pred_class_final = (y_pred_final > 0.5).astype(int)

# Create the confusion matrix
cm_final = confusion_matrix(y, y_pred_class_final)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_final, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Final Model)')
plt.show()
```

True Positives (TP): 378 → The model correctly predicted "1" when the actual class was "1."

True Negatives (TN): 82 → The model correctly predicted "0" when the actual class was "0."

False Positives (FP): 97 → The model incorrectly predicted "1" when the actual class was "0" (Type I error).

False Negatives (FN): 7 → The model incorrectly predicted "0" when the actual class was "1" (Type II error).

```python
# prompt: Extract AUC(Area under curve )

from sklearn.metrics import roc_auc_score

# Assuming 'final_model' and 'final_X' are from the previous code
# and 'y' is your target variable

# Predict probabilities for the positive class
y_pred_prob = final_model.predict(final_X)

# Calculate AUC
auc = roc_auc_score(y, y_pred_prob)

print(f"AUC: {auc}")
```

AUC of 0.756 indicates good model performance, meaning it correctly distinguishes loan approvals ~75.6% of the time.
