```python
# prompt: Check for ouliers on data

import pandas as pd
import numpy as np

# Assuming 'data' DataFrame is already loaded as shown in the previous code

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers


# Example usage: Detecting outliers in the 'Calories' column
outliers_calories = detect_outliers_iqr(data, 'Big Mac Price')
print("Outliers in 'Big Mac Price':\n", outliers_calories)


# Example usage: Detecting outliers in all numerical columns
numerical_cols = data.select_dtypes(include=np.number).columns
for col in numerical_cols:
    outliers = detect_outliers_iqr(data, col)
    if not outliers.empty:  # Print only if there are outliers
        print(f"\nOutliers in '{col}':\n{outliers}")

# You can adapt this function to different outlier detection methods, such as Z-score.
```

```python
# prompt: let the depndent variable y be New Hourly wage and independent variable x be Big Mac price

import pandas as pd
import statsmodels.api as sm

# Assuming 'data' DataFrame is already loaded and preprocessed

# Define the dependent and independent variables
X = data['Big Mac Price']
y = data['New Hourly Wage']

```

```python
# prompt: Extract Simple linear regression on x and y

# Add a constant to the independent variable (required for statsmodels)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
```

```python
# now we check for tye assumptions
# Residual
# homoscedasticity
```

```python
# prompt: check whether residuals are normally distributed using histogram with line

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import norm

# ... (Your existing code) ...

# Assuming 'model' is your fitted OLS model from the previous code

# Get the residuals
residuals = model.resid

# Create a histogram of the residuals
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=30, density=True, alpha=0.6, color='b', label='Residuals')

# Overlay a normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, residuals.mean(), residuals.std())
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')

plt.xlabel('Residuals')
plt.ylabel('Density')
plt.title('Histogram of Residuals with Normal Distribution Overlay')
plt.legend(loc='best')
plt.show()
```

```python
# prompt: check for homoscedasticity

import matplotlib.pyplot as plt
# Check for homoscedasticity (constant variance of errors)
plt.figure(figsize=(8, 6))
plt.scatter(model.fittedvalues, residuals)  # Plot predicted values vs. residuals
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at 0
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

# Breusch-Pagan test for heteroscedasticity
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(residuals, model.model.exog)
print("Breusch-Pagan Test:", bp_test)
labels = ['Lagrange multiplier statistic', 'p-value',
          'f-value', 'f p-value']
print(dict(zip(labels, bp_test)))
alpha = 0.05
if bp_test[1] > alpha:
    print("Homoscedasticity is present (fail to reject H0)")
else:
    print("Heteroscedasticity is present (reject H0)")
```
