```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

# Define the categorical columns
categorical_cols = ['gender', 'relevent_experience', 'enrolled_university', 'education_level',
                    'major_discipline', 'experience', 'company_size', 'company_type']

# Apply label encoding to each categorical column
for col in categorical_cols:
    data[col] = le.fit_transform(data[col])

# Display the updated DataFrame
data.head()
```

Categorical columns such as gender, relevent_experience, enrolled_university, education_level, major_discipline, experience, company_size, and company_type were successfully converted into numerical values using LabelEncoder.

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Define the numerical columns to scale
numerical_cols = ['city_development_index', 'training_hours', 'experience']

# Fit and transform the numerical columns
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Display the updated DataFrame
data.head()
```

The numerical columns â€” city_development_index, training_hours, and experience â€” have been standardized using StandardScaler so that there will not be any dominance of any of the columns

```python
import matplotlib.pyplot as plt
# Check for outliers using box plots
plt.figure(figsize=(15, 12))
for i, col in enumerate(data.columns):
    plt.subplot(4, 3, i + 1)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

we are checking for outliers in the columns of the data
significant amount of outliers are found in "training_hours"

```python
import numpy as np
# Calculate Q1, Q3, and IQR for 'training_hours'
Q1 = data['training_hours'].quantile(0.25)
Q3 = data['training_hours'].quantile(0.75)
IQR = Q3 - Q1

# Define the bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and replace outliers with the bounds
data['training_hours'] = np.where(data['training_hours'] < lower_bound, lower_bound, data['training_hours'])
data['training_hours'] = np.where(data['training_hours'] > upper_bound, upper_bound, data['training_hours'])
```

Here we are treating the outlier in the training_hour column using Inter Quartile range

```python
import matplotlib.pyplot as plt
# Check for outliers using box plots after outlier treatment
plt.figure(figsize=(15, 12))
for i, col in enumerate(data.columns):
    plt.subplot(4, 3, i + 1)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

Making the box plot again and now the outlier in the training_hour column is treated successfully
